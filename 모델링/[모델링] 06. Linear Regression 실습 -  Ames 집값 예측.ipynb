````python
# ë°ì´í„°ì…‹ì„ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤.
import pandas as pd

homes = pd.read_csv('https://ds-lecture-data.s3.ap-northeast-2.amazonaws.com/ames_home/ames_home_subset.csv')
assert homes.shape == (2904, 47)
````

### **Train/Validation/Test ì…‹ ë¶„í• **

- **Train**: 2006ë…„ë¶€í„° 2008ë…„ê¹Œì§€ íŒ”ë¦° ì£¼íƒë“¤ (1,920ê°œ)

- **Validation**: 2009ë…„ì— íŒ”ë¦° ì£¼íƒë“¤. (644ê°œ)

````python
# ì—°ë„ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ì„¸ ê°œì˜ ë°ì´í„°ì…‹ìœ¼ë¡œ ë¶„í• í•©ë‹ˆë‹¤.

Train_df = homes[homes['YrSold'] <= 2008].reset_index(drop=True)
Val_df = homes[homes['YrSold'] == 2009].reset_index(drop=True)
Test_df = homes[homes['YrSold'] == 2010].reset_index(drop=True)

print('Train: ',len(Train_df.index))
print('Validation: ',len(Val_df.index))
print('Test: ',len(Test_df.index))
````

````python
# Featureê°’ê³¼ Targetê°’ì„ ë¶„ë¦¬í•©ë‹ˆë‹¤.

X_train = Train_df.drop(['YrSold'],axis=1)
X_val = Val_df.drop(['YrSold'],axis=1)
X_test = Test_df.drop(['YrSold'],axis=1)

y_train = Train_df['YrSold']
y_val = Val_df['YrSold']
y_test = Test_df['YrSold']
````

### **One-Hot Encoding**

````python
# ì›í•«ì¸ì½”ë”©ì„ ì‹¤í–‰í•©ë‹ˆë‹¤.
from category_encoders import OneHotEncoder

encoder = OneHotEncoder(use_cat_names = True)
X_train = encoder.fit_transform(X_train)
X_val = encoder.transform(X_val)
X_test = encoder.transform(X_test)
````

### **Linear Regression**

````python
# leanear íšŒê·€ ëª¨ë¸ì„ ìƒì„± í›„ í•™ìŠµí•©ë‹ˆë‹¤.

from sklearn.linear_model import LinearRegression

lir = LinearRegression()
lir.fit(X_train, y_train)
````
